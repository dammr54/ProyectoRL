```python
import matplotlib.pyplot as plt
import pickle
import numpy as np

def load(filename):
    file = open(filename, "rb")
    data = pickle.load(file)
    file.close()
    return data

def dump(filename, obj):
    file = open(filename, "wb")
    pickle.dump(obj, file)
    file.close()
```

```python
lista_distancias_stable = load("all_distances_step_50000.pickle")
lista_distancia_propia = load("all_distances_step_31000.pickle")
```

```python
fig, ax = plt.subplots(dpi=350, figsize=[6, 3])
ax.plot(lista_distancias_stable[0:31000], label="stable-baselines3")
ax.plot(lista_distancia_propia, label="own implementation")
ax.set_yscale("log")
ax.set_xlabel("Steps")
ax.set_ylabel("Distance to Target (m)")
ax.set_title("Distance to Target vs Steps")
ax.legend()
```
